```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('../appliance_energy.csv')
df
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
sns.boxplot(df)
plt.show()
```

```python
sns.histplot(df, x='Temperature (C)')
plt.xlabel('Temperature')
plt.ylabel('Energy Consumption')
plt.show()
```

```python
print('Highest temperature:', df['Temperature (C)'].max())
print('Lowest temperature:', df['Temperature (C)'].min())
```

```python
print('Highest Consumption:', df['Energy Consumption (kWh)'].max())
print('Lowest Consumption:', df['Energy Consumption (kWh)'].min())
```

```python
y = df[['Energy Consumption (kWh)']]
x = df[['Temperature (C)']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=45, test_size=0.2)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
y_pred = linear_model.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color='Green')
plt.title('Energy Consumption Prediction using Simple Linear Regression')
plt.xlabel('Temperature')
plt.ylabel('Energy Consumption (kWh)')
plt.show()
```
