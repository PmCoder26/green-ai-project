```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'jun']
energy_consumption = [1200, 1300, 1100, 1500, 1400, 1600]
```

```python
df = pd.DataFrame({
    "Months": months,
    "Energy Consumption": energy_consumption
})
df
```

```python
plt.plot(df['Months'], df['Energy Consumption'], marker='*', linestyle='--', color='darkblue')
plt.show()
```

```python
plt.bar(df['Months'], df['Energy Consumption'], color='lightblue')
plt.title('Energy Consumption over 6 months')
plt.xlabel("Months")
plt.ylabel("Energy Consumption")
plt.show()
```

```python
# '%1.2f%%' is a format string that displays a floating-point number with 2 decimal places followed by a percent sign.
# wedgeprops={'edgecolor': 'white', 'linewidth': 1} sets the border color and thickness for each pie slice to create visible separation between them.
plt.pie(df['Energy Consumption'], labels=df['Months'], autopct='%1.2f%%', wedgeprops={'edgecolor': 'white', 'linewidth': 1})
plt.title('Energy Consumption by months')
plt.show()
```
